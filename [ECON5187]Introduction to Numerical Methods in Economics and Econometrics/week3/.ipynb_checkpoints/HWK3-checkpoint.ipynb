## * Problem set1: Draw Data Using Inverse Transform Sampling

A half-normal distribution is often denoted as $N^+(0, \sigma^2)$, which means it is a distribution obtained by truncating a normal distribution $N(0, \sigma^2)$ at the point of `0` and keeping only the positive part (thus the "+" sign) of it. Since it cuts the **underlying normal distribution** into half, it is called a **half-normal distribution**. As you can see from the following graph, the density of $N^+(0, \sigma^2)$ is twice as higher than that of the $N(0, \sigma^2)$.

This homework problem asks you to draw samples from the half-normal distribution $N^+(0, 2)$. It can be done in several ways.

````julia
# draw 5000 random values from N+(0, 2) and its theoretical distribution

using Distributions, LaTeXStrings, StatsPlots

d = truncated(Normal(0, sqrt(2)); lower=0)
data1 = rand(d, 5000)

plot(d, linewidth=3, label="theoretical", xlim=[-1.5, 5], ylim=[0, 0.7])
histogram!(data1, normalize=true, bin=200, label="empirical")
vline!([0], color=:black, linewidth=2)
````

````julia
# use `quantile` to do the same thing as above cell

using Distributions, LaTeXStrings, StatsPlots

d = truncated(Normal(0, sqrt(2)); lower=0)
data2 = quantile(d, rand(5000))

plot(d, linewidth=3, label="theoretical", xlim=[-1.5, 5], ylim=[0, 0.7])
histogram!(data2, normalize=true, bin=200, label="empirical")
vline!([0], color=:black, linewidth=2)
````

````julia
# For the explanation of the logic behind that line of code, please refer to the following Markdown cell.

data3 = quantile(Normal(0, sqrt(2)), 0.5 * rand(5000) .+ 0.5)

plot(d, linewidth=3, label="theoretical", xlim=[-1.5, 5], ylim=[0, 0.7])
histogram!(data3, normalize=true, bin=200, label="empirical")
vline!([0], color=:black, linewidth=2)
````

#### The explanation of the logic behind `quantile(Normal(0, sqrt(2)), 0.5 * rand(5000) .+ 0.5)`
Define a random variable $X^{\prime}$ following the distribution $N^+(0, \sigma^2)$ and its cumulative distribution function (CDF) $F_{X^{\prime}}(x^{\prime})$. Also, define a random variable $X$ following the distribution $N(0, \sigma^2)$ and its CDF $F_{X}(x)$.

When $x^{\prime} \geq 0$, we have:
$$
\begin{align*}
F_{X^{\prime}}(x^{\prime}) &= P(X^{\prime} \leq x^{\prime}) \\
&= P(F_{X^{\prime}}(X^{\prime}) \leq F_{X^{\prime}}(x^{\prime})) \\
&= P(2(F_{X}(X^{\prime})-0.5) \leq 2(F_{x}(x^{\prime})-0.5)) \\
&= P(F_{X}(X^{\prime}) \leq F_{X}(x^{\prime})) \\
&= 2(F_{X}(x^{\prime})-0.5)
\end{align*}
$$

Let $Y$ be a new random variable that is equal to $F_X(X^{\prime})$ and $y$ be a constant that equals $F_X(x^{\prime})$.
$$
\begin{align*}
& P(F_{X}(X^{\prime}) \leq F_{X}(x^{\prime})) = 2(F_{X}(x^{\prime})-0.5) \\
&P(Y \leq y) = 2(y - 0.5) = \frac{y - 0.5}{0.5} 
\end{align*}
$$

We know that $y$ should be in the range of $[0.5, \ 1]$ according to the constraint of $x^{\prime}$ and the definition of CDF. It is not difficult to see that $Y\sim U(0.5, 1)$. Therefore, we can apply the probability integral transformation to obtain the sampling result from a half-normal distribution.

````julia
# compare the three datasets I generated to the theoretical values
println("mean(theoretical) = $(mean(d))")
println("standard deviation(theoretical) = $(sqrt(var(d)))")
println("---------------------------------------------------")
println("mean(data1) = $(mean(data1))")
println("standard deviation(data1) = $(sqrt(var(data1)))")
println("---------------------------------------------------")
println("mean(data2) = $(mean(data2))")
println("standard deviation(data2) = $(sqrt(var(data2)))")
println("---------------------------------------------------")
println("mean(data3) = $(mean(data3))")
println("standard deviation(data3) = $(sqrt(var(data3)))")
````

## * Problem set2: In the class, we showed how to transform values from a normal random variable $X \sim F_X(X)$ to values of an exponential random variable $Z \sim F_Z(Z)$. We used a three-step procedure to do this, and we drew graphs to have a visual confirmation on the transformation. This homework problem asks you to modify the graphs we used in class and observe the relationship between the sample size (`n` in the example) and the approximation of the distribution.

### Take a random sample $X$ of `n` measurements from $N(1,2)$ (note: the variance is 2), and draw a histogram of $X$ and compare it to the true $N(1,2)$ density (by overlaying the density to the histogram). Add a slider of $n$ from 100 to 20000. 
- Hint: You need to use packages of `Distributions`, `Plots`, `Interact`, `WebIO`, `StatsPlots`.
- Hint: Use `@manipulate` with a `for` loop.

````julia
using Distributions, Interact, StatsPlots, Plots, LaTeXStrings

d = Normal(1, sqrt(2))

@manipulate for n in (100:100:20000)
    data = rand(d, n)
    plot(d, linewidth=3, label="theoretical", xlim=[-4, 6], ylim=[-0.1, 0.4])
    histogram!(data, normalize=true, bin=200, label="empirical")
    vline!([1], color=:black, linewidth=2)
end
````

### Continue from the above: Transform $X$ to $Y \sim U(0,1)$, and draw a histogram of $Y$ and compare it to a $U(0,1)$. Overlaying the histogram by the density plot of $U(0,1)$.
Hint: You will need `cdf()` function to transform `X` to `Y`.

````julia
using Distributions, Interact, StatsPlots, Plots, LaTeXStrings

d = Normal(1, sqrt(2))

@manipulate for n in (100:100:20000)
    data = cdf(d, rand(d, n))
    plot(Uniform(0, 1), linewidth=3, label="theoretical", xlim=[-0.1, 1.1], ylim=[0, 2])
    histogram!(data, normalize=true, bin=200, label="empirical")
end
````

### Continue from the above: Transform $Y$ to $Z ~ Exp(2)$ (exponential distribution, where mean=standard deviation=2) using an appropriate quantile function, and  draw a histogram of $Z$ and compare it to a Exp$(2)$. Overlay the histogram by a density plot of $Exp(2)$.

Note: If the histogram resembles the density plot, congratulations! You've successfully transformed normally distributed data to exponential distribution data.

````julia
using Distributions, Interact, StatsPlots, Plots, LaTeXStrings

d = Normal(1, sqrt(2))

@manipulate for n in (100:100:20000)
    data = quantile(Exponential(2), cdf(d, rand(d, n)))
    plot(Exponential(2), linewidth=3, label="theoretical", xlim=[-1, 10], ylim=[0, 0.7])
    histogram!(data, normalize=true, bin=200, label="empirical")
end
````

##  * Problem set3: The 2nd step in the above transformation relies on PIT (probability integral transformation) which we introduced in the class. However, PIT works best only if the CDF we used in the transformation is close to the true CDF of $X$. This problem asks you to examine what would happen if the CDF used in the transformation is not the actual CDF of $X$.

### Take a random sample $X$ of `n` measurements from $N(1,2)$. Transform $X$ to $Y \sim U(0,1)$ using the CDF of $N(1,1)$. Draw a histogram of $Y$ and overlay it with the density plot of $U(0,1)$. Remember to add a slider of `n`. According to your visual, does $Y$ look like uniformly distributed?

````julia
using Distributions, Interact, StatsPlots, Plots, LaTeXStrings

d = Normal(1, sqrt(2))

@manipulate for n in (100:100:20000)
    data = cdf(Normal(1, 1), rand(d, n))
    plot(Uniform(0, 1), linewidth=3, label="theoretical", xlim=[-0.1, 1.1], ylim=[0, 2])
    histogram!(data, normalize=true, bin=200, label="empirical")
end

# the histogram does not look like a U(0, 1) distribution at all
````

### Continue from the above: Transform $Y$ to $Z ~ Exp(2)$ (exponential distribution, where mean=standard deviation=2) using an appropriate quantile function, and  draw a histogram of $Z$ and compare it to a Exp$(2)$. Overlay the histogram by a density plot of $Exp(2)$.

Question: Does the histogram match the density plot well? Why not?

````julia
using Distributions, Interact, StatsPlots, Plots, LaTeXStrings

d = Normal(1, sqrt(2))

@manipulate for n in (100:100:20000)
    data = quantile(Exponential(2), cdf(Normal(1, 1), rand(d, n)))
    plot(Exponential(2), linewidth=3, label="theoretical", xlim=[-1, 15], ylim=[0, 1])
    histogram!(data, normalize=true, bin=200, label="empirical")
end

# It does not match the density plot because `cdf(Normal(1, 1), rand(Normal(1, sqrt(2)), n))` is not a U(0, 1) distribution
````
